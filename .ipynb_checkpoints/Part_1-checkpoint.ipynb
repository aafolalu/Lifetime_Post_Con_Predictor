```python
# IMPORTED LIBRARIES
import pandas as pd
from collections import Counter
```

```python
# LOAD DATASET INTO DATAFRAME
data = pd.read_csv("Facebook_metrics/dataset_Facebook.csv",delimiter=";")
data.head()
```

# Question 1

```python
# DISTINCT TYPES OF POST
data["Type"].unique()
```

```python
# MOST COMMON TYPE OF POST
data["Type"].mode()
```

```python
# MOST COMMON TYPE OF POST WITH FREQUENCY
Counter(data["Type"]).most_common(4)
```

# Question 2

```python
# AVERAGE (MEDIAN) PAGE TOTAL LIKES PER TYPE
data.groupby(['Type'], as_index=False).median().groupby('Type')['share'].median()
```

# Question 3

```python
# AVERAGE (MEDIAN) NUMBER OF SHARES PER PAID
data.groupby(['Paid'], as_index=False).median().groupby('Paid')['share'].median()
```

# Consumption per user

```python
# EXTRACT CONSUMPTION PER USER COLUMN
d = (data["Lifetime Post Consumptions"])/(data["Lifetime Post Consumers"])
data.insert(12, "Consumption per User", d)
data.head()
```

# 

```python
# AVERAGE (MEDIAN) CONSUMPTION PER USER FOR EACH TYPE
data.groupby(['Type'], as_index=False).median().groupby('Type')['Consumption per User'].median()
```

```python
# AVERAGE (MEDIAN) CONSUMPTION PER USER FOR EACH MONTH
data.groupby(['Post Month'], as_index=False).median().groupby('Post Month')['Consumption per User'].median()
```

```python
# AVERAGE (MEDIAN) CONSUMPTION PER USER FOR EACH TYPE
data.groupby(['Type'], as_index=False).median().groupby('Type')['Lifetime Post Consumers'].median()
```

```python
# AVERAGE (MEDIAN) CONSUMPTION PER USER FOR EACH MONTH
data.groupby(['Post Month'], as_index=False).median().groupby('Post Month')['Lifetime Post Consumers'].median()
```

```python
# SELECT PHOTO TYPE POSTS
photo_type = data[data['Type'] == "Photo"]
print(photo_type)
```

```python
# AVERAGE (MEDIAN) CONSUMPTION PER USER FOR EACH MONTH
data.groupby(['Post Month'], as_index=False).skew().groupby('Post Month')['Consumption per User'].skew()
```
