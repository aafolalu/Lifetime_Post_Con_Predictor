# Feature Engineering

```python
# UNIQUE VALUES IN 'TYPE' COLUMN
data.Type.unique()
```

```python
# ENCODING THE 'TYPE' COLUMN
t = []
for n, i in enumerate(data.Type):
    if i == 'Photo':
        i = 1.0
        t.append(i)
    elif i == 'Status':
        i = 2.0
        t.append(i)
    elif i == 'Link':
        i = 3.0
        t.append(i)
    elif i == 'Video':
        i = 4.0
        t.append(i)

data["Type2"] = t
data.head()
```

```python
# REMOVING OUTLIERS
outlier = np.percentile(data["Page total likes"],90)
data = data[data["Page total likes"]<outlier]
print(data.shape)
data.head()
```

```python
# SELECT FEATURES
X = data[['Page total likes', 'Category', 'Post Month', 'Post Weekday', 'Post Hour', 'Paid', 'Type2']] # all features
# X = data[['Page total likes', 'Category', 'Post Month', 'Post Weekday', 'Paid', 'Type2']] # six features
y = data[['Lifetime Post Consumers']]
X.head()
# y.head()
```

```python
# SCALE FEATURES BETWEEN 0-1
scaler = MinMaxScaler()
col = X.columns
X = pd.DataFrame(scaler.fit_transform(X[::]), columns=col)
X.head()
```

```python
# OPTIMUM FEATURE SELECTION
estimator = SVR(kernel="linear", C=1)
rfe = RFE(estimator, n_features_to_select=6)
r = rfe.fit(X, y)
rank = rfe.ranking_
print(rank)
```

```python
# LOAD DATA INTO DMATRIX
data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
# SPLIT DATA INTO TRAIN AND TEST
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
y_test.head()
```

# XGBoost Regressor and Grid Search 

```python
# GRID SEARCH TO FIND BEST PARAMETERS FOR XGBOOST REG
param = {
 'n_estimators':[i for i in range(188,199)],
 'max_depth':[i for i in range(3,7)],
 'min_child_weight':[i for i in range(2,6)],
 'colsample_bytree':[0.95,0.96,0.97],
 'colsample_bylevel':[0.95,0.96,0.97]
}

gsearch1 = GridSearchCV(
    estimator = xgb.XGBRegressor(objective= 'reg:linear', subsample = 1.0,
                                 seed=1, n_stopping_rounds=10, learning_rate = 0.01), 
    param_grid = param, 
    scoring='mean_absolute_error',
    cv=10,
    verbose = 1)

gsearch1.fit(X_train, y_train)
```

```python
# BEST GRID SEARCH SCORE FOR XGBOOST REG
gsearch1.best_score_
```

```python
# BEST GRID SEARCH PARAMETERS FOR XGBOOST REG
gsearch1.best_params_
```

```python
# INSTANTIATE XGBOOST REGRESSOR
xg_reg = xgb.XGBRegressor(objective ='reg:linear', 
                          subsample = 1.0, seed=1, 
                          n_stopping_rounds=10, 
                          learning_rate = 0.01,
                          n_estimators = 188, 
                          max_depth = 4, 
                          min_child_weight = 4, 
                          colsample_bytree = 0.95, 
                          colsample_bylevel = 0.95)
```

```python
# # FIT XGBOOST TO TRAINING SET
# xg_reg.fit(X_train, y_train)
# # PREDICT WITH TEST SET
# pred = xg_reg.predict(X_test)
```

```python
# 10-FOLD CROSS VALIDATION FOR MODEL EVALUATION
scores = cv.cross_val_score(xg_reg, X_train, y_train, cv=10,  scoring='mean_absolute_error')
print("MAE: %0.2f (+/- %0.2f)" % (abs(scores.mean()), scores.std()))
```
